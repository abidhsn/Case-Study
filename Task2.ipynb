```python
# Create features dataframe on address level

def extract_features_addresses(df):
    transaction_count = df.groupby('address')['address'].count().rename("total_trns")
    incoming_transaction_count = df[df["incoming_transaction"] == True].groupby('address')['address'].count().rename("in_trns")
    outgoing_transaction_count = df[df["incoming_transaction"] == False].groupby('address')['address'].count().rename("out_trns")
    average_eth = df.groupby('address')['values_int'].mean().rename("avg_eth")
    total_eth = df.groupby('address')['values_int'].sum().rename("total_eth")
    average_eth_sent = df[df["incoming_transaction"] == False].groupby('address')['values_int'].mean().rename("avg_eth_out")
    average_eth_received = df[df["incoming_transaction"] == True].groupby('address')['values_int'].mean().rename("avg_eth_in")
    total_eth_sent = df[df["incoming_transaction"] == False].groupby('address')['values_int'].sum().rename("total_eth_out")
    total_eth_received = df[df["incoming_transaction"] == True].groupby('address')['values_int'].sum().rename("total_eth_in")
    std_eth_received = df[df["incoming_transaction"] == True].groupby('address')['values_int'].std().rename("std_eth_in")
    std_eth_sent = df[df["incoming_transaction"] == False].groupby('address')['values_int'].std().rename("std_eth_out")
    average_gasprice_received = df[df["incoming_transaction"] == True].groupby('address')['gas'].mean().rename("avg_gasprice_in")
    average_gasprice_sent = df[df["incoming_transaction"] == False].groupby('address')['gas'].mean().rename("avg_gasprice_out")
    duration_activity = (df.groupby('address')['timeStamp'].max() - df.groupby('address')['timeStamp'].min()).rename("time_active")
    features_df = pd.concat([transaction_count,incoming_transaction_count, outgoing_transaction_count, average_eth,
                            total_eth, average_eth_sent, average_eth_received, total_eth_sent, total_eth_received, std_eth_received, 
                            std_eth_sent, average_gasprice_received, average_gasprice_sent, duration_activity],axis=1)

    return features_df
```
